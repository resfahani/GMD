# Import data

```python
path = "/Users/rezad.d.esfahani/Desktop/Project2/code/GFZ_PC/Main_Code/Data_Gen/"

def import_data(path):
    
    data_all = pd.read_csv(path + 'ESM_flatfile_FAS.csv',sep = ';', low_memory=False)

    FAS_labels = [x for x in data_all.columns if '_F' in x]
    data_dropna = data_all.dropna(subset = FAS_labels)

    d = data_dropna[data_dropna.Mw.notna()]
    d = d[d.epi_dist.notna()]
    d = d[d.ev_depth_km.notna()]

    M = d.Mw.values
    
    Depi = np.sqrt(d.epi_dist.values**2 + d.ev_depth_km.values**2)

    mag_dis = []
    mag_dis.append(M)
    mag_dis.append(Depi)
    mag_dis = np.asarray(mag_dis)
    
    FAS_U_labels = [x for x in d.columns if 'U_F' in x]
    FAS_V_labels = [x for x in d.columns if 'V_F' in x]
    FAS_W_labels = [x for x in d.columns if 'W_F' in x]
    data_3comp   = np.asarray(d[FAS_U_labels + FAS_W_labels + FAS_V_labels])


    freq  = [float(x[3:].replace("_", ".")) for x in FAS_U_labels]
    freq = np.asarray(freq)
    n_freq       = len(freq)

    ## add small value to avoid log(0)
    eps = 1e-15
    #data_3comp = np.sqrt(data_3comp[:, :n_freq]**2 + data_3comp[:, n_freq:(2*n_freq)]**2) #+ data_3comp[:, 2*n_freq:(3*n_freq)]**2)
    data_3comp = np.sqrt( data_3comp[:, 2*n_freq:(3*n_freq)]**2 )
    data_3comp  += eps

    ## take log10
    data_main   = np.log(data_3comp)

        
    return data_main[:,15:99], mag_dis.T, freq[15:99]


#def data_filter(data, mag_dis):
    
```

```python
# Dividing data to training, validation, and test dataset

train_ratio = 0.75
validation_ratio = 0.15
test_ratio = 0.10


md_train, md_val, data_train, data_val = train_test_split(mag_dis, data, test_size=1 - train_ratio, random_state=1) 
md_val, md_test, data_val, data_test = train_test_split(md_val, data_val, test_size=test_ratio/(test_ratio + validation_ratio), random_state=1) 


batch_size = 64
learning_rate = 1e-3
earlystoping = 10
numofepochs = 1000
reg = 1e-3
```

```python

## Inital 
ls1_1 = []
ls1_2 = []

for i in range(1,10):
    latent_space = i
    ls1 = []
    ls2 = []
    
    print("latentspace : " + str(i))

    for ii in range(20):

        model_AE, model_enc, model_dec = AEModel(latent_space, original_dim=n_freq , alpha = reg)

        
        cc = optimizers.Adam(learning_rate = learning_rate)
        model_AE.compile(loss='mean_squared_error', optimizer=cc)
        #model_AE.summary()
        
        es = EarlyStopping(monitor='val_loss', verbose = False, patience = earlystoping) 
        
        history = model_AE.fit(data_train, data_train, shuffle = True, epochs = numofepochs, verbose = False,
                                  batch_size=batch_size, validation_data = (data_val, data_val),
                                  callbacks=[History(),es])


        axout = model_AE.predict(data, batch_size = batch_size, verbose = False)
        
        #ls1.append(np.mean(np.abs(axout-data),0))
        ls1.append(np.mean(np.abs(axout-data)))
        ls2.append((np.abs(axout-data)))
        
    #print("latentspace : " + str(i) ) #+ ", iteration : " + str(ii))
    
    ls1_1.append(ls1)
    ls1_2.append(np.asarray(ls2).reshape(-1,))
    
    
```
